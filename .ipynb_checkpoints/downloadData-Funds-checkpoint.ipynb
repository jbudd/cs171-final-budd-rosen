```python
%matplotlib inline

import time
import json
import requests
import math

import matplotlib.pyplot as plt

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl

dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def fetch_all_data_generator(data):
    for company in data:
        yield requests.get(entity_api(api_key, 'financial-organization', company['permalink'])).text
```

```python
def fetch_all_data(data, start, end, verbose=0):
    if verbose >= 1:
        print time.strftime('%H:%M:%S')
    data = data[start:end]
    outfile = open(set_version('allFirms', 'json'), 'w')
    outfile.write('[')
    for company_data in fetch_all_data_generator(data): 
        try:
            company = json.loads(company_data)
            company_out = {
                            'name':company['name'],
                            'permalink':company['permalink'],
                            'crunchbase_url':company['crunchbase_url'],
                            'website':company['homepage_url'],
                            'founded_year':company['founded_year'],
                            'description':company['description'],
                            'investments':company['investments'],
                            'funds':company['funds']
                          }
            json.dump(company_out, outfile)
            outfile.write(',')
        except:
            if verbose >= 2:
                print company_data        
    outfile.seek(-1, 1)
    outfile.truncate()
    outfile.write(']')
    outfile.close()
    if verbose >= 1:
        print time.strftime('%H:%M:%S')
    return
```

```python
# this cell loads the local JSON file of all the data on financial organizations we downloaded
# the data is current as of _____

json_data = open('allFirms_2014-04-07-01-06-44.json')
results = json.load(json_data)
json_data.close() 
```

```python
def check_none(f, verbose=0):
    if f['permalink'] == None or f['crunchbase_url'] == None or f['name'] == None or len(f['investments']) < 1:
        if verbose > 0:
            print '\n%s' % f
        return False
    else:
        return True
```

```python
results = [f for f in results if check_none(f, verbose=0) == True]
print 'We retained %d data point after filtering/cleaning' % len(results)

# print results
```

```python
json_data = open('allCompaniesList_2014-04-02-12-47-25.json')
all_companies_data = json.load(json_data)
json_data.close()
```

```python
# adds total number of investments for each firm

for n, i in enumerate(results):
    results[n]['total'] = len(i['investments'])
```

```python
all_num = [] 
for i in results:
    all_num.append(i['total'])
```

```python
plt.hist(all_num, round(math.sqrt(len(all_num)) / 100) * 100, log=True)
plt.xlabel('Number of Investments')
plt.xlim(0, max(all_num))
plt.ylabel('Frequency')
plt.title('Number of Investments per Firm')
remove_border()
```

```python
# all_num_200 = [c for c in all_num if c >= 200]
# print len(all_num_200)
```

```python
# keeps only frims with more than x total investments

all_num = sorted(all_num, reverse=True)
all_num = all_num[0:50]
cut_off_num = all_num[-1]

print cut_off_num

results_out = []
for i in results:
    if i['total'] >= cut_off_num:
        results_out.append(i)

# print len(results_out)
```

```python
all_companies_codes = {}
for c in all_companies_data:
    all_companies_codes[c['permalink']] = (c['category_code'] if c['category_code'] != None else 'other')
                           
# print all_companies_codes
```

```python
# adds category codes to all companies

for n, i in enumerate(results_out):
    for nn, fr in enumerate(i['investments']):
        results_out[n]['investments'][nn]['funding_round']['company']['category_code'] = (all_companies_codes[fr['funding_round']['company']['permalink']] if fr['funding_round']['company']['permalink'] in all_companies_codes else 'other')
        
        
```

```python
# finds all co-occurences

for n, i in enumerate(results_out):
    results_out[n]['cooc'] = {}
    for fr in i['investments']:
        for ii in results_out:
            for frfr in ii['investments']:
                if fr['funding_round'] == frfr['funding_round']:
                    results_out[n]['cooc'][ii['permalink']] = (i['cooc'][ii['permalink']] + 1 if ii['permalink'] in i['cooc'] else 1)
```

```python
# breaks down investments by catagory

for n, i in enumerate(results_out):
    results_out[n]['cat'] = {}
    for fr in i['investments']:
        results_out[n]['cat'][fr['funding_round']['company']['category_code']] = (results_out[n]['cat'][fr['funding_round']['company']['category_code']] + 1 if fr['funding_round']['company']['category_code'] in results_out[n]['cat'] else 1)
        
        
```

```python
# breaks down investments by round
all_types = []
rounds = ['seed', 'a', 'b', 'c', 'd+', 'other']
for n, i in enumerate(results_out):
    results_out[n]['round'] = {}
    for r in rounds:
        results_out[n]['round'][r] = 0
    for fr in i['investments']:
        all_types.append(fr['funding_round']['round_code'])
        if  any(fr['funding_round']['round_code'] == r for r in rounds):  
            results_out[n]['round'][fr['funding_round']['round_code']] += 1 
        else:
            if len(fr['funding_round']['round_code']) == 1:
                results_out[n]['round']['d+'] += 1
            else:
                results_out[n]['round']['other'] += 1
all_types = set(all_types)
        
```

```python
# breaks down investments by year

for n, i in enumerate(results_out):
    results_out[n]['time'] = {}
    for fr in i['investments']:
        results_out[n]['time'][fr['funding_round']['funded_year']] = (results_out[n]['time'][fr['funding_round']['funded_year']] + 1 if fr['funding_round']['funded_year'] in results_out[n]['time'] else 1)
        
         
```

```python
totals = [c['total'] for c in results_out]
firms = [c['name'] for c in results_out]
plt.bar(range(len(results_out)), totals, align='center')
plt.xlim(-1,len(results_out))
plt.xticks(range(len(results_out)), firms, rotation='vertical')
plt.title('Investments by Firm')
remove_border()
```

```python
def second_largest(numbers):
    m1, m2 = None, None
    for x in numbers:
        if x >= m1:
            m1, m2 = x, m1
        elif x > m2:
            m2 = x
    return m2
```

```python
plt.bar(range(len(results_out[num]['cooc'])), results_out[num]['cooc'].values(), align='center')
plt.xticks(range(len(results_out[num]['cooc'])), results_out[num]['cooc'].keys(), rotation='vertical')
plt.xlim(-1,len(results_out[num]['cooc']))
# the tallest bar (the firms co-investments with themselves) is truncated
plt.ylim(0, (second_largest(results_out[num]['cooc'].values())+10))
plt.title('%s\'s Co-Investments' % results_out[num]['name'])
remove_border()
```
